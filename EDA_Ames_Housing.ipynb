## Outlier Treatment:

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.boxplot(y=df['SalePrice'])
plt.title('SalePrice - Before Treatment')

plt.subplot(1,2,2)
sns.histplot(df['SalePrice'], kde=True)
plt.title('SalePrice Distribution - Before')

plt.tight_layout()
plt.show()
```

```python
## Step 1 - Calculate IQR boundaries

Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Q1: {Q1} \nQ3: {Q3} \nIQR: {IQR}\n")

print(f"Lower bound: {lower}")
print(f"Upper bound: {upper}\n")

print(f"Outliers count: {((df['SalePrice'] < lower) | (df['SalePrice'] > upper)).sum()}")
```

```python
# Cap outliers at boundaries
df['SalePrice'] = df['SalePrice'].clip(lower=lower, upper=upper)

# Visualise after treatment
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.boxplot(y=df['SalePrice'])
plt.title('SalePrice - After Treatment')

plt.subplot(1,2,2)
sns.histplot(df['SalePrice'], kde=True)
plt.title('SalePrice Distribution - After')

plt.tight_layout()
plt.show()
```

#### Outlier Treatment - Summary

Method used: IQR Capping

* Q1: 129,500 |    Q3: 213,500 |    IQR: 84,000
* Lower bound: 3,500 | Upper bound: 339,500

* Outliers detected: 137 (4.6% of data)

* Treatment: Capped outliers at boundary values rather than removing them — since these are real transactions, we preserve the rows but limit extreme values.

* Result: No more extreme outliers. Spike at 339,500 is expected — capped values accumulate at the boundary.

# Day 4: Advanced Feature Engineering

## Encoding and its types:

```python
# Run this first
print(df['Sale Type'].nunique())
print(df['Sale Type'].value_counts())
```

### 1. Ordinal Encoding on Exter Qual:

```python
print(df['Exter Qual'].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the order manually — lowest to highest
order = [ ['Fa', 'TA', 'Gd', 'Ex'] ]

oe = OrdinalEncoder(categories = order)
df['Exter Qual Encoded'] = oe.fit_transform(df[['Exter Qual']])

print(df[['Exter Qual', 'Exter Qual Encoded']].value_counts().reset_index().sort_values('Exter Qual Encoded'))
```

### 2. One Hot Encoding on Neighborhood:

```python
print(df['Neighborhood'].unique())
print(df['Neighborhood'].nunique())
```

```python
# Apply N-1 Dummy Encoding on Neighborhood column
# drop_first=True drops one category to avoid redundancy

neighborhood_dummies = pd.get_dummies(df['Neighborhood'], drop_first = True)


# Verify how many columns were created vs unique values
print(f"Original Unique values: {df['Neighborhood'].nunique()}")
print(f"New columns created: {neighborhood_dummies.shape[1]}")


# Join the new dummy columns back to the main dataframe
df = pd.concat([df, neighborhood_dummies], axis = 1)


# Drop the original Neighborhood column as it has been replaced by encoded columns
df.drop(columns = ['Neighborhood'], inplace = True)

# Verify new shape of dataframe
print(df.shape)
```

### 3. Frequency Encoding on Sale Type:

```python
# Calculate frequency of each category
# value_counts() counts occurrences of each Sale Type
# dividing by len(df) converts counts to proportions (0 to 1)
freq = df['Sale Type'].value_counts() / len(df)


# Map frequency back to original column
# .map(freq) replaces each category name with its frequency value
# creates a new numeric column 'Sale Type Encoded'
df['Sale Type Encoded'] = df['Sale Type'].map(freq)


# Verify the encoding by selecting both original and encoded columns together
# .reset_index() converts result to a clean dataframe
# sorting by encoded value to see highest to lowest frequency
print(df[['Sale Type', 'Sale Type Encoded']]
      .value_counts()
      .reset_index()
      .sort_values('Sale Type Encoded', ascending=False))
```

#### Encoding Summary

1. Ordinal Encoding → Exter Qual
   Order defined: Fa(0) < TA(1) < Gd(2) < Ex(3)
   Used when categories have a natural ranking

2. N-1 Dummy Encoding → Neighborhood
   28 unique values → 27 new columns created
   Original column dropped after encoding
   Used for nominal categories with no order

3. Frequency Encoding → Sale Type
   Each category replaced with its proportion in data
   WD most frequent (0.865), VWD least (0.0003)
   Used when one category dominates heavily

## Feature Scaling:
Making all numeric columns comparable in range so no single column dominates just because of its unit or scale

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Scaling on Total SF
# initialise the MinMaxScaler
mms = MinMaxScaler()

# fit and transform Total SF column
# fit learns the min and max values
# transform applies the formula (X - min) / (max - min)
df['Total SF Scaled'] = mms.fit_transform(df[['Total SF']])

# Standardisation on Gr Liv Area
# initialise the StandardScaler
ss = StandardScaler()

# fit and transform Gr Liv Area column
# fit learns the mean and std values
# transform applies the formula (X - mean) / std
df['Gr Liv Area Scaled'] = ss.fit_transform(df[['Gr Liv Area']])

# Verify both scaled columns
print(df[['Total SF', 'Total SF Scaled', 
          'Gr Liv Area', 'Gr Liv Area Scaled']].head())
```

```python
# check actual min and max of scaled columns
print("Total SF Scaled:")
print(f"  Min: {df['Total SF Scaled'].min():.2f}")
print(f"  Max: {df['Total SF Scaled'].max():.2f}")

print("\nGr Liv Area Scaled:")
print(f"  Min: {df['Gr Liv Area Scaled'].min():.2f}")
print(f"  Max: {df['Gr Liv Area Scaled'].max():.2f}")
```

## Data Transformation:
This is about fixing the shape of your data distributions — mainly making skewed columns more normal.

#### Check skewness of numeric columns

```python
# Step 1 : Check skewness of numeric columns

numeric_cols = df.select_dtypes(include='number').columns
skewness = df[numeric_cols].skew().sort_values(ascending=False)
print(skewness.head(106))  # top 10 most skewed
```

#### Skewness Insights:
* Values > 1 → highly right skewed (long tail on right) → needs log transform
* Values < -1 → highly left skewed → needs sqrt or other fix
* Values between -1 and 1 → acceptable, no transformation needed
* SalePrice, Lot Area, Gr Liv Area are typically > 1 in this dataset

### 1. Log Transformation

```python
# Step 2 — Apply Log Transformation on a skewed column (e.g. SalePrice and Lot Area are classically skewed):

# Log Transform on SalePrice (right skewed)
# Why log1p? It handles zeros safely. log(0) = error, log1p(0) = 0

df['SalePrice Log'] = np.log1p(df['SalePrice'])

# Compare before vs after
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['SalePrice'].hist(ax=axes[0], bins=50)
axes[0].set_title('SalePrice - Original (Skewed)')

df['SalePrice Log'].hist(ax=axes[1], bins=50)
axes[1].set_title('SalePrice - After Log Transform')

plt.tight_layout()
plt.show()
```

#### Insights:

* Before → histogram looks like a mountain pushed to the left with a long right tail
* After → histogram should look more bell-shaped / symmetric
* That's the goal — transformation worked if the skew reduced

### 2. Square Root Transformation:

```python
# Step 3 — Square Root Transform (for moderately skewed columns like Lot Area):

df['Lot Area Sqrt'] = np.sqrt(df['Lot Area'])

print(f"Lot Area skew before: {df['Lot Area'].skew():.2f}")
print(f"Lot Area skew after:  {df['Lot Area Sqrt'].skew():.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['Lot Area'].hist(ax=axes[0], bins=50)
axes[0].set_title(f"Lot Area - Original (Skew: {df['Lot Area'].skew():.2f})")

df['Lot Area Sqrt'].hist(ax=axes[1], bins=50)
axes[1].set_title(f"Lot Area - After Sqrt (Skew: {df['Lot Area Sqrt'].skew():.2f})")

plt.tight_layout()
plt.show()
```

#### Insights:

* Before → likely ~12+ (very high skew, lots of small plots + few huge ones)
* After sqrt → should drop to ~2–4 (reduced but may not be perfect)
* Plot both histograms to visually confirm it got more symmetric

### 3. BoxCox Transformation

* Automatically finds the best power (λ) to normalize your data
* Limitation: only works on positive values (no zeros, no negatives)
* Use when: data is skewed + all values are strictly > 0

```python
from scipy.stats import boxcox
df['Lot Area BoxCox'], lambda_val = boxcox(df['Lot Area'] + 1)
print(f"Best lambda found: {lambda_val:.2f}")
```

### 4. Yeo-Johnson Transformation

* Same idea as Box-Cox but works on zeros and negative values too
* Safer, more flexible — preferred in most real projects

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df['Gr Liv Area YJ'] = pt.fit_transform(df[['Gr Liv Area']])
```

```python
# Compare skewness before vs after
print(f"Original skew: {df['Gr Liv Area'].skew():.3f}")
print(f"After YJ skew: {df['Gr Liv Area YJ'].skew():.3f}")

# Visual comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['Gr Liv Area'].hist(ax=axes[0], bins=50)
axes[0].set_title('Original - Gr Liv Area')

df['Gr Liv Area YJ'].hist(ax=axes[1], bins=50)
axes[1].set_title('After Yeo-Johnson Transform')
plt.show()
```
